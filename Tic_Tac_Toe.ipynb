# TIC TAC TOE GAME


```python
#Basic Method
# Initialize board
board = [' ' for _ in range(9)]  # 3x3 grid
current_player = 'X'

# Function to print the board
def print_board():
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--")
    print(f"{board[6]} | {board[7]} | {board[8]}")
    print("\n")

# Function to check if there's a winner
def check_win(player):
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # columns
        [0, 4, 8], [2, 4, 6]  # diagonals
    ]
    for condition in win_conditions:
        if board[condition[0]] == board[condition[1]] == board[condition[2]] == player:
            return True
    return False

# Function to check if the game is a draw
def check_draw():
    if ' ' not in board:  # No empty spaces left
        return True
    return False

# Function to handle player's move
def make_move(player):
    while True:
        try:
            move = int(input(f"Player {player}, enter your move (1-9): ")) - 1
            if board[move] == ' ':
                board[move] = player
                break
            else:
                print("That spot is already taken.")
        except (ValueError, IndexError):
            print("Invalid input. Please enter a number between 1 and 9.")

# Main game loop
def play_game():
    global current_player
    print("Welcome to Tic-Tac-Toe!")
    print_board()

    while True:
        make_move(current_player)
        print_board()

        # Check for win
        if check_win(current_player):
            print(f"Player {current_player} wins!")
            break

        # Check for draw
        if check_draw():
            print("The game is a draw!")
            break

        # Switch player
        current_player = 'O' if current_player == 'X' else 'X'

# Start the game
if __name__ == "__main__":
    play_game()
```

```python
import math

def check_win(player):
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
        [0, 4, 8], [2, 4, 6]  # Diagonals
    ]
    return any(board[c[0]] == board[c[1]] == board[c[2]] == player for c in win_conditions)

def is_draw():
    return ' ' not in board

def minimax(depth, is_maximizing, alpha, beta):
    if check_win('X'):
        return 1
    if check_win('O'):
        return -1
    if is_draw():
        return 0

    if is_maximizing:
        max_eval = -math.inf
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'X'
                eval_score = minimax(depth + 1, False, alpha, beta)
                board[i] = ' '
                max_eval = max(max_eval, eval_score)
                alpha = max(alpha, eval_score)
                if beta <= alpha:
                    break
        return max_eval
    else:
        min_eval = math.inf
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'O'
                eval_score = minimax(depth + 1, True, alpha, beta)
                board[i] = ' '
                min_eval = min(min_eval, eval_score)
                beta = min(beta, eval_score)
                if beta <= alpha:
                    break
        return min_eval

def best_move():
    best_score = -math.inf
    move = -1
    for i in range(9):
        if board[i] == ' ':
            board[i] = 'X'
            score = minimax(0, False, -math.inf, math.inf)
            board[i] = ' '
            if score > best_score:
                best_score = score
                move = i
    board[move] = 'X'
    print(f"AI chooses position {move + 1}")

def player_move():
    while True:
        try:
            move = int(input("Your move (1-9): ")) - 1
            if board[move] == ' ':
                board[move] = 'O'
                break
            else:
                print("That spot is already taken.")
        except (ValueError, IndexError):
            print("Invalid input. Enter a number between 1-9.")

def print_board():
    print("\n")
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("---+---+---")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("---+---+---")
    print(f" {board[6]} | {board[7]} | {board[8]} ")
    print("\n")

def play_game(starting_player):
    global board
    board = [' ' for _ in range(9)]
    print_board()
    ai_turn = (starting_player == 'AI')
    while True:
        if ai_turn:
            best_move()
            print_board()
            if check_win('X'):
                print("AI wins!")
                break
        else:
            player_move()
            print_board()
            if check_win('O'):
                print("You win!")
                break

        if is_draw():
            print("It's a draw!")
            break
        ai_turn = not ai_turn

def main():
    starting_player = 'AI'
    while True:
        print("New Game!")
        play_game(starting_player)
        starting_player = 'Human' if starting_player == 'AI' else 'AI'
        if input("Do you want to play again? (y/n): ").lower() != 'y':
            break

print("AI is ready to play against you!")
main()
```

# Q Learning


```python
import random
import numpy as np
import pickle
from collections import deque

class TicTacToe:
    def __init__(self):
        self.board = np.array([' '] * 9)
        self.ai_x = 'X'
        self.ai_o = 'O'
        self.current_player = 'X'
        self.winner = None

    def reset(self):
        self.board.fill(' ')
        self.current_player = 'X'
        self.winner = None
        return self.get_state()

    def get_state(self):
        return tuple(self.board)

    def available_actions(self):
        return np.where(self.board == ' ')[0].tolist()

    def step(self, action):
        if self.board[action] != ' ':
            return tuple(self.board), 0

        self.board[action] = self.current_player
        reward = self.check_winner()
        self.current_player = 'O' if self.current_player == 'X' else 'X'
        return tuple(self.board), reward

    def check_winner(self):
        win_conditions = [
            [0, 1, 2], [3, 4, 5], [6, 7, 8],
            [0, 3, 6], [1, 4, 7], [2, 5, 8],
            [0, 4, 8], [2, 4, 6]
        ]
        for condition in win_conditions:
            if self.board[condition[0]] == self.board[condition[1]] == self.board[condition[2]] and self.board[condition[0]] != ' ':
                self.winner = self.board[condition[0]]
                return 10 if self.winner == 'X' else -10
        if ' ' not in self.board:
            return 2
        return 0

class TicTacToeQLearning:
    def __init__(self, alpha=0.05, gamma=0.95, epsilon=1.0, epsilon_decay=0.999, min_epsilon=0.01):
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.min_epsilon = min_epsilon
        self.q_table_x = {}
        self.q_table_o = {}
        self.experience_replay = deque(maxlen=10000)
        self.game = TicTacToe()

    def choose_move(self, state, player):
        available_actions = self.game.available_actions()
        if random.uniform(0, 1) < self.epsilon:
            return random.choice(available_actions)
        else:
            if player == 'X':
                return max(available_actions, key=lambda action: self.q_table_x.get((state, action), 0))
            else:
                return min(available_actions, key=lambda action: self.q_table_o.get((state, action), 0))

    def update_q_values(self, state, action, reward, next_state, done, player):
        if done:
            best_next_q = 0
        else:
            available_actions = self.game.available_actions()
            if available_actions:
                best_next_q = max(self.q_table_x.get((next_state, a), 0) for a in available_actions)
            else:
                best_next_q = 0

        # If AI X (maximizer) played, update Q-table X
        if player == 'X':
            self.q_table_x[(state, action)] = self.q_table_x.get((state, action), 0) + self.alpha * (
                reward + self.gamma * best_next_q - self.q_table_x.get((state, action), 0)
            )
        # If AI O (minimizer) played, update Q-table O
        else:
            self.q_table_o[(state, action)] = self.q_table_o.get((state, action), 0) - self.alpha * (
                reward + self.gamma * best_next_q - self.q_table_o.get((state, action), 0)
            )

    def store_experience(self, state, action, reward, next_state, done):
        self.experience_replay.append((state, action, reward, next_state, done))

    def replay_experience(self):
        if len(self.experience_replay) < 10:
            return
        batch = random.sample(self.experience_replay, min(64, len(self.experience_replay)))
        for state, action, reward, next_state, done in batch:
            self.update_q_values(state, action, reward, next_state, done, player='X')

    def train(self, episodes, save_interval=10000):
        for episode in range(episodes):
            state = self.game.reset()
            while ' ' in state:
                player = self.game.current_player
                action = self.choose_move(state, player)
                next_state, reward = self.game.step(action)
                self.update_q_values(state, action, reward, next_state, ' ' not in next_state, player)
                self.store_experience(state, action, reward, next_state, ' ' not in next_state)
                state = next_state

            if self.epsilon > self.min_epsilon:
                self.epsilon *= self.epsilon_decay

            self.replay_experience()

            if episode % save_interval == 0:
                self.save_q_table(f"q_table_{episode}.pkl")

    def save_q_table(self, filename="q_table.pkl"):
        with open(filename, "wb") as f:
            pickle.dump({"X": self.q_table_x, "O": self.q_table_o}, f)

# **Hyperparameter Tuning**
alpha_values = [0.01, 0.03, 0.05]
gamma_values = [0.99, 0.95]
epsilon_decay_values = [0.999, 0.9995, 0.9999]

best_win_rate = 0
best_hyperparams = {}

for alpha in alpha_values:
    for gamma in gamma_values:
        for epsilon_decay in epsilon_decay_values:
            ai = TicTacToeQLearning(alpha=alpha, gamma=gamma, epsilon=1.0, epsilon_decay=epsilon_decay)
            ai.train(50000)

            wins = 0
            for _ in range(30000):
                ai.game.reset()
                while ' ' in ai.game.get_state():
                    action = ai.choose_move(ai.game.get_state(), player='X')
                    ai.game.step(action)
                if ai.game.check_winner() == 10:  # Reward AI X winning
                    wins += 1

            win_rate = wins / 30000

            print(f"Alpha: {alpha}, Gamma: {gamma}, Epsilon Decay: {epsilon_decay}, Win Rate: {win_rate:.3f}")

            if win_rate > best_win_rate:
                best_win_rate = win_rate
                best_hyperparams = {"alpha": alpha, "gamma": gamma, "epsilon_decay": epsilon_decay}
                print("New Best:", best_hyperparams)

print("Best Hyperparameters:", best_hyperparams)
print("Best Win Rate:", best_win_rate)

# **Final Save**
ai.save_q_table("final_q_table.pkl")
print("✅ Training Complete! Final Q-table saved.")
```
